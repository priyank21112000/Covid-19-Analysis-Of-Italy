# Relating data with Scatterplots

```python
sns.relplot(x="TotalPositiveCases",y="Recovered",data=data)
```

# Conclusion:More people got recovered with increasing cases i.e. Recovery Rate increased

```python
sns.relplot(x="TotalPositiveCases",y="Deaths",data=data)
```

# Conclusion:Total positive cases and deaths increased exponentially

```python
sns.relplot(y="TotalPositiveCases",x="TestsPerformed",data=data)
```

# Conclusion:From 0 to 200000 tests performed there were less positive cases as compared when tests performed were 200000 to 400000

```python
sns.relplot(x="HospitalizedPatients",y="TotalPositiveCases",data=data)
```

```python
sns.relplot(x="TotalPositiveCases",y='HospitalizedPatients',hue='Recovered',data=data)
```

```python
sns.pairplot(data) #Complete plot of all the values
```
